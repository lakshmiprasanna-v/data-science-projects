```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "/content/anthropometric_data.csv"  # Change this if needed
df = pd.read_csv(file_path)
```

```python
# Display first few rows
print("First 5 rows of dataset:")
print(df.head())
```

```python
print("\nMissing values in dataset:")
print(df.isnull().sum())
```

```python
# 2. Descriptive statistics
print("\nDescriptive statistics:")
print(df.describe())
```

```python


# Define figure size
plt.figure(figsize=(12, 8))

# Create histograms with improved styling
df.hist(bins=30, figsize=(12, 8), edgecolor='black', color='skyblue')

# Improve title visibility
plt.suptitle("Distribution of Anthropometric Measurements", fontsize=16, fontweight='bold')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python

# 4. Box plots for outlier detection
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plot of Anthropometric Measurements")
plt.xticks(rotation=45)
plt.show()
```

```python



# 5. Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Anthropometric Measurements")
plt.show()
```

```python
# 6. Calculate percentiles for adaptive furniture design
percentiles = df.quantile([0.05, 0.50, 0.95])
print("\nKey Percentiles (5th, 50th, 95th):")
print(percentiles)
```

```python
# 7. Adaptive Furniture Design Recommendations
# Example: Seat height should accommodate popliteal height
popliteal_height_mean = df["popliteal_height"].mean()
seat_height_recommendation = (popliteal_height_mean - 20, popliteal_height_mean + 20)
print(f"\nRecommended Seat Height Range: {seat_height_recommendation[0]:.2f} mm to {seat_height_recommendation[1]:.2f} mm")
```

```python
# 8. Recommended desk height based on elbow rest height
elbow_rest_mean = df["elbow_rest_height"].mean()
desk_height_recommendation = (elbow_rest_mean + 50, elbow_rest_mean + 100)

print(f"Recommended Desk Height Range: {desk_height_recommendation[0]:.2f} mm to {desk_height_recommendation[1]:.2f} mm")
```

```python
shoulder_height_mean = df["shoulder_height"].mean()
backrest_height_recommendation = (shoulder_height_mean - 100, shoulder_height_mean)

print(f"Recommended Backrest Height Range: {backrest_height_recommendation[0]:.2f} mm to {backrest_height_recommendation[1]:.2f} mm")
```

```python
# 10. Save statistical summary to a CSV file
summary_csv = "anthropometric_summary.csv"
df.describe().to_csv(summary_csv)
print(f"\nStatistical summary saved to {summary_csv}")
```
